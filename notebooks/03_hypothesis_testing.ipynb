```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("../data/salaries.csv")
```

**Proposed question:** Do fully remote employees earn higher salaries (in USD) than on-site or hybrid employees?

**Hypotheses:**

* $H_0$: The mean salary is the same across all remote ratios.
* $H_1$: At least one remote ratio has a different mean salary.

```python
groups = [g["salary_in_usd"] for _, g in df.groupby("remote_ratio")]

f_stat, p_value = stats.f_oneway(*groups)
print(
    f"F-statistic = {f_stat}",
    f"\np-value = {p_value}",
)
```

Since $\text{pâ€“value} < 0.05$, we reject $H_0$.

This also matches with our previous boxplot observation.

**Conclusion:** Mean salaries are not equal across remote ratios.

**Interpretation:** At least one remote ratios (on-site, hybrid, fully remote) has a significant different salary.
